## Reverse Geocoding

### [Nominatim](https://wiki.openstreetmap.org/wiki/Nominatim)

```python
import requests
io = []
isIndoor = []
for i in range(0, data['lat'].size):
    params = (
        ('format', 'jsonv2'),
        ('lat', data['lat'][i]),
        ('lon', data['lon'][i]),
        ('zoom', '18'),
        ('addressdetails', '1'),
    )

    r = requests.get('https://nominatim.openstreetmap.org/reverse', params=params)

    io.append(r.json()['category'])
    
    location_type = r.json()['category']
    
    if location_type == "building" or location_type == "amenity":
        isIndoor.append(1);
    else:
        isIndoor.append(0);

    print("{} : {}".format(data['idx'][i], io[i]))
```

```python
data['location_type'] = io
data['isIndoor'] = isIndoor
data.to_csv('data_io.csv', index=False)
```
